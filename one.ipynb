```python
average_salary = df['Salary'].mean()
```

```python
average_salary.astype('int')
```

```python
above_avg_salary = df[df['Salary'] >= average_salary]
```

```python
len_of_above_avg_salary = len(above_avg_salary)
```

```python
print("Length of above average salary is:", len_of_above_avg_salary)
```

```python
avg_salary_age = above_avg_salary['Age'].value_counts()
```

```python
avg_salary_dep = above_avg_salary['Department_Region'].value_counts()
```

```python
avg_salary_age
```

```python
avg_salary_dep.head(10)
```

```python
below_avg_salary = df[df['Salary'] <= average_salary]
```

```python
len_of_below_avg_salary = len(below_avg_salary)
```

```python
print("Length of below average salary is:", len_of_below_avg_salary)
```

```python
avg_below_salary_age = below_avg_salary['Age'].value_counts()
```

```python
avg_below_salary_age
```

```python
avg_below_salary_dep = below_avg_salary['Department_Region'].value_counts()
```

```python
avg_below_salary_dep.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
print("Average Salary is:", average_salary.astype('int'))
print("Length of above average salary is:", len_of_above_avg_salary) 
print("Length of below average salary is:", len_of_below_avg_salary)

fig, ax = plt.subplots(figsize=(8,5))

above_below = [len_of_above_avg_salary, len_of_below_avg_salary]

ax.bar(
    ['Above Average', 'Below Average'],
    above_below,
    color=['mediumseagreen', 'lightcoral']
)

ax.set_title('Above vs Below Average Salary', fontsize=15)
ax.set_ylabel('Count')
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create full name
low_7_salary['Name'] = low_7_salary['First_Name'] + ' ' + low_7_salary['Last_Name']

fig, ax = plt.subplots(figsize=(12, 6))

colors = plt.cm.Reds(range(70, 250, 25))

bars = ax.barh(
    low_7_salary['Name'],
    low_7_salary['Salary'],
    color=colors,
    edgecolor='black'
)

ax.set_title("Top 7 low salary earners")
ax.set_xlabel("Salary")
ax.set_ylabel("Name")
```

```python
# Create full name
high_7_salary['Name'] = high_7_salary['First_Name'] + ' ' + high_7_salary['Last_Name']

fig, ax = plt.subplots(figsize=(12, 6))

colors = plt.cm.Greens(range(70, 250, 25))

bars = ax.barh(
    high_7_salary['Name'],
    high_7_salary['Salary'],
    color=colors,
    edgecolor='black'
)

ax.set_title("Top 7 high salary earners")
ax.set_xlabel("Salary")
ax.set_ylabel("Name")
```
